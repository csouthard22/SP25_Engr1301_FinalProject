```python
import matplotlib.pyplot as plt
from IPython.display import clear_output
import math
import numpy
```

# 1

```python
x = [i for i in range(-330, 331)]
y = [757.7 - 127.7*math.cosh(xval/127.7) for xval in x]

fig, ax = plt.subplots()
ax.plot(x, y)
ax.axis([-400,400,-100,700])
ax.set_aspect('equal')
ax.grid(True)
plt.show()
```

# 2

```python
x = [i for i in numpy.arange(-4*math.pi,4*math.pi+.001, .001)]
y = [math.sin(xval)/xval for xval in x]

fig, ax = plt.subplots()
ax.plot(x, y)
plt.axhline(y=0, color='red')
plt.axvline(x=0, color='red')
ax.axis([-4*math.pi,4*math.pi,-1.5,1.5])
ax.grid(True)
plt.show()
```

# 3

```python
x = [i for i in numpy.arange(-1,1.001,.0001)]
y = [math.sin(1/xval) for xval in x]

fig, ax = plt.subplots()
ax.plot(x, y)
ax.axis([-1,1,-1.5,1.5])
ax.grid(True)
plt.title('Plot of Sin(1/x)')
plt.show()
```

# 4

```python
def approx(n):
    sum = 0
    for i in range(0,n+1):
        if i % 2 == 0:
            sum += 1/(2*i+1)
        else:
            sum -= 1/(2*i+1)
    return sum*4

x = [i for i in range(0,100)]
y = [approx(xval) for xval in x]

fig, ax = plt.subplots()
ax.plot(x, y)
plt.xlabel('Number of terms')
plt.ylabel('Value of Approximation')
plt.title('Approximation of Pi: 4(1/1 - 1/3 + 1/5 - 1/7 + ...)')
plt.show()
```

# 5

```python
D = 137.51
def r(k): return math.sqrt(k)
def theta(k): return (math.pi * D * k)/180

time = [i for i in range(1,1001)]
x = [r(t) * math.sin(theta(t)) for t in time]
y = [r(t) * math.cos(theta(t)) for t in time]

fig, ax = plt.subplots()
ax.scatter(x, y, s=10, facecolors='none', edgecolors='green')
ax.set_aspect('equal')
ax.axis('off')
plt.show()
```

# 6

```python
file = open("DJIA_monthly_open_Feb_1992_Feb_2025.txt")
values = [float(line) for line in file.readlines()]

MONTHLY_RATE = 0.0065
val = values[0]
predictedValues = [val]

for i in range(1, len(values)):
    val = val * (1 + MONTHLY_RATE)
    predictedValues.append(val)

fig, ax = plt.subplots()
plt.title('Actual vs Estimated Dow Jones Return')
plt.xlabel('Months Starting Feb 1992')
plt.ylabel('Monthly Opening Price')
plt.plot(values, label='Actual')
plt.plot(predictedValues, color='orange', label='Estimated')
plt.legend()
ax.text(0,35000, f'Estimated Monthly Rate of Return: {MONTHLY_RATE*100}%')
ax.text(0,32000, f'Estimated Annual Rate of Return: {MONTHLY_RATE*12*100}%')
plt.show()
```
