```python
import matplotlib.pyplot as plt
import numpy
import math
```

# Line Plotting Challenges

```python
x = [0,1,2,5,6,6,8]
y = [0,3,0,7,7,0,3]

fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
x_val = [i for i in range(-2,3)]
y = [2*x**2 - x**4 for x in x_val]

fig, ax = plt.subplots()
ax.plot(x_val,y)
plt.grid()
plt.show()
```

```python
x_val = [i for i in numpy.arange(-2.1,1,.1)]
y = [x + x**2 + x**3 - x**4 - 2.7*x**5 - x**6 for x in x_val]

fig, ax = plt.subplots()
ax.plot(x_val,y)
plt.grid()
plt.show()
```

```python
x_val = [i for i in numpy.arange(-1,1.1,.1)]
y = [1/x for x in x_val]

fig, ax = plt.subplots()
ax.plot(x_val,y)
plt.grid()
plt.show()
```

```python
x_val = [i for i in numpy.arange(-2,2.1,.1)]
y = [math.sin(x) for x in x_val]

fig, ax = plt.subplots()
ax.plot(x_val,y)
plt.grid()
plt.show()
```

```python
x_val = [i for i in numpy.arange(-2,2.001,.001)]
y = [math.sin(1/x) for x in x_val]

fig, ax = plt.subplots()
ax.plot(x_val,y)
plt.grid()
plt.show()
```

```python
x_val = [i for i in numpy.arange(-1.25,1.26,.01)]
y1 = [math.sin(x) for x in x_val]
y2 = [math.cos(x) for x in x_val]
y3 = [math.tan(x) for x in x_val]

fig, ax = plt.subplots()
ax.plot(x_val,y1)
ax.plot(x_val,y2)
ax.plot(x_val,y3)
plt.grid()
plt.show()
```

# Plot Styling Challenges

```python
x_val = [i for i in range(-3,4)]
y = [x**2 for x in x_val]

fig, ax = plt.subplots()
ax.plot(x_val,y, color='purple', marker='d')
ax.axis('equal')
plt.title('Graphing x squared')
plt.xlabel('x')
plt.ylabel('x^2')
plt.grid()
plt.show()
```

# Parametric Plotting Challenges

```python
time = [i for i in numpy.arange(0, 2 * math.pi + .1, .1)]
x = [math.cos(t) for t in time]
y = [math.sin(t) for t in time]

fig, ax = plt.subplots()
ax.plot(x,y)
ax.axis('equal')
plt.title('Unit Circle')
plt.grid()
plt.show()
```

```python
time = [i for i in numpy.arange(0, 5 * math.pi + .1, .1)]
x = [t*math.cos(t) for t in time]
y = [t*math.sin(t) for t in time]

fig, ax = plt.subplots()
ax.plot(x,y)
ax.axis('equal')
plt.grid()
plt.show()
```

```python
time = [i for i in numpy.arange(0, 2 * math.pi + .1, .1)]
y = [math.sin(t) for t in time]

fig, ax = plt.subplots()
ax.plot(time,y)
plt.grid()
plt.show()
```

```python
time = [i for i in numpy.arange(0, 2 * math.pi + .1, .1)]
y = [math.sin(t) for t in time]

fig, ax = plt.subplots()
ax.plot(y,time)
plt.grid()
plt.show()
```

```python
k = 0.9
l = 0.3

time = [i for i in numpy.arange(0, 100.1, .1)]
x = [(1-k) * math.cos(t) + l*k*math.cos(((1-k)/k)*t) for t in time]
y = [(1-k) * math.sin(t) + l*k*math.sin(((1-k)/k)*t) for t in time]

fig, ax = plt.subplots()
ax.plot(x,y)
plt.grid()
plt.show()
```

# Figure Plotting Challenges

```python
time = [i for i in numpy.arange(0, 2*math.pi+1, 2*math.pi/3)]
x = [math.cos(t) for t in time]
y = [math.sin(t) for t in time]

fig, ax = plt.subplots()
ax.plot(x,y)
plt.axis('equal')
plt.grid()
plt.show()
```

```python
time = [i for i in numpy.arange(0, 2*math.pi+1, 2*math.pi/4)]
x = [math.cos(t) for t in time]
y = [math.sin(t) for t in time]

fig, ax = plt.subplots()
ax.plot(x,y)
plt.axis('equal')
plt.grid()
plt.show()
```

```python
# Original coordinates
time = [i for i in numpy.arange(0, 2*math.pi+1, 2*math.pi/4)]
x = [math.cos(t) for t in time]
y = [math.sin(t) for t in time]

# Rotation by 45 degrees
theta = math.radians(45)
x_rot = [x[i] * math.cos(theta) - y[i] * math.sin(theta) for i in range(len(x))]
y_rot = [x[i] * math.sin(theta) + y[i] * math.cos(theta) for i in range(len(x))]

fig, ax = plt.subplots()
ax.plot(x_rot, y_rot)
plt.axis('equal')
plt.grid()
plt.show()
```
