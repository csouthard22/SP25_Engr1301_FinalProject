# Structured Data

```python
import json
import pprint
file = open("prize.json", 'r')
prizes = json.load(file)
```

```python
print(f'Bob Dylan was awarded a prize for {prizes['prizes'][50]['category']}.')
```

```python
print("Economic prizes:")
print("--------------")

for prize in prizes['prizes']:
    if prize.get('category', '').lower() == 'economics':
        year = prize.get('year')
        
        laureates_info = []
        for laureate in prize['laureates']:
            name = laureate.get('firstname', '') + ' ' + laureate.get('surname', '')
            motivation = laureate.get('motivation')
            laureates_info.append(f"Laureate: {name.strip()}, Motivation: {motivation}")
        
        print(f"Year: {year}")
        if laureates_info:
            for info in laureates_info:
                print(f"  {info}")
        else:
            print("  No laureates information available")
        print("--------------")
```

```python
not_awarded_count = 0

print("Years and categories where prizes were not awarded:")
print("-------------------------------------------------")

for prize in prizes['prizes']:
    overall_motivation = prize.get('overallMotivation', '')
    if overall_motivation and 'No Nobel Prize' in overall_motivation:
        year = prize.get('year', 'Unknown')
        category = prize.get('category', 'Unknown')
        
        print(f"Year: {year}, Category: {category}")
        not_awarded_count += 1

print("-------------------------------------------------")
print(f"Total count of category-years not awarded: {not_awarded_count}")
```

```python
search_surname = input("Enter a surname to search for: ").lower()
print(f"\nSearching for Nobel laureates with surname containing '{search_surname}'...\n")

found_laureates = []

for prize in prizes['prizes']:
    year = prize.get('year', 'Unknown')
    category = prize.get('category', 'Unknown')
    
    if 'laureates' in prize:
        for laureate in prize['laureates']:
            surname = laureate.get('surname', '').lower()
            
            if search_surname in surname:
                firstname = laureate.get('firstname', '')
                full_name = f"{firstname} {laureate.get('surname', '')}"
                motivation = laureate.get('motivation', 'No motivation provided')
                
                found_laureates.append({
                    'name': full_name.strip(),
                    'year': year,
                    'category': category,
                    'motivation': motivation
                })

if found_laureates:
    print(f"Found {len(found_laureates)} matching laureate(s):\n")
    print("-" * 80)
    
    for i, laureate in enumerate(found_laureates, 1):
        print(f"Match #{i}:")
        print(f"Name: {laureate['name']}")
        print(f"Year: {laureate['year']}")
        print(f"Category: {laureate['category'].title()}")
        print(f"Motivation: {laureate['motivation']}")
        print("-" * 80)
else:
    print(f"No Nobel laureates found with surname containing '{search_surname}'.")
```
