# Data Quality and Cleaning

```python
file = open('war_and_peace.txt', 'r')
text = file.read()
file.close()

word_count = {}
text = text.lower()
text = text.replace('!', ' ')
text = text.replace('.', ' ')
text = text.replace('?', ' ')
text = text.replace(',', ' ')
text = text.replace(':', ' ')
text = text.replace(';', ' ')
text = text.replace('(', ' ')
text = text.replace(')', ' ')
text = text.replace('\'', ' ')
text = text.replace('-', ' ')
text = text.replace('$', ' ')
text = text.replace('*', ' ')
text = text.replace('—', ' ')
text = text.replace('‘', ' ')
text = text.replace('“', ' ')
text = text.replace('”', ' ')

for word in text.split():
    word_count[word] = word_count.get(word, 0) + 1


pprint.pprint(word_count)
print( word_count[ 'happy' ] )
print( word_count[ 'sad' ] )
print( word_count[ 'war' ] )
print( word_count[ 'peace' ] )
```

# Second Analysis - Better Sentiment Analysis

```python
def count_by_list(word_count_dict, word_list):
    """ Counts how many times all of the words in the
        word_list occur in the word_count_dict, and
        returns a single value.
        
        Parameters:
            word_count_dict: a dictionary of word counts where
            keys are words and their values are integer counts
            of how many times they appear

            word_list: a list of words to aggregate counts for

        Returns:
            The sum of word counts for all words in the word_list
    """
    sentiment = 0

    for word in word_list:
        if word in word_count_dict:
            sentiment += word_count_dict[word]
    return sentiment
    

word_list = ['war', 'wartime', 'warfare', 'warlike', 'warrior', 'warriors', 'wars']

count_by_list(word_count, word_list)
```
