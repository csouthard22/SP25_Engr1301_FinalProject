## Plotting the surface z = sin(x) + cos(y)

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Make data.
X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(X, Y)
Z = np.sin(X) + np.cos(Y)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)

plt.show()
```

# Demo 2: Python as a Data Processor

SoFIFA is a dataset of soccer player stats derived from the FIFA line of games published by Electronic Arts.

## Load SoFIFA data

```python
import pandas as pd

# Replace 'path_to_file.csv' with the actual path to the file
data = pd.read_csv('sofifa_2024.csv')

# Show the available columns
print(data.columns)
```

## Tallest and shortest soccer players

```python
heights = data.height_cm 

tallest = heights.max()
shortest = heights.min()

cm_in_foot = 30.5

print(f"The tallest player is {tallest} cm and the shortest is {shortest} cm")
print(f"That's also {tallest/cm_in_foot:.2f} feet and {shortest/cm_in_foot:.2f} feet")
```

## Relationship between weight and acceleration

```python
import matplotlib.pyplot as plt
import numpy as np

# Remove any rows with empty values
cleaned_data = cleaned_data = data.dropna(subset=['weight_kg', 'acceleration'])

X = cleaned_data.weight_kg
Y = cleaned_data.acceleration

# Scatterplot with marker size 1
plt.scatter(X, Y, 1)

# Calculate a best-fit linear trendline, with 1 being 1st degree polynomial (i.e. linear)
Z = np.polyfit(X, Y, 1)
P = np.poly1d(Z)
plt.plot(X,P(X))

# Add labels
plt.xlabel('Weight (kg)', fontsize=12)  # X-axis label
plt.ylabel('Acceleration', fontsize=12)  # Y-axis label
plt.title('Player Weight vs. Acceleration', fontsize=14, fontweight='bold')  # Title

# Add grid (optional)
plt.grid(True)

plt.show()
```

## Relationship between overall rating and acceleration

```python
import matplotlib.pyplot as plt
import numpy as np

# Remove any rows with empty values
cleaned_data = cleaned_data = data.dropna(subset=['overall_rating', 'acceleration'])

# Remove any rows with empty values
X = cleaned_data.overall_rating
Y = cleaned_data.acceleration

# Scatterplot with marker size 1
plt.scatter(X, Y, 1)

# Calculate a best-fit linear trendline, with 1 being 1st degree polynomial (i.e. linear)
Z = np.polyfit(X, Y, 1)
P = np.poly1d(Z)
plt.plot(X,P(X))

# Add labels
plt.xlabel('Overall Rating', fontsize=12)  # X-axis label
plt.ylabel('Acceleration', fontsize=12)  # Y-axis label
plt.title('Overall Rating vs. Acceleration', fontsize=14, fontweight='bold')  # Title

# Add grid (optional)
plt.grid(True)

plt.show()
```

## Play a game of guess the random number

```python
import random

print("Welcome to the 'Guess the Number' game!")
print("I am thinking of a number between 1 and 10. Can you guess it?")

# Generate a random number between 1 and 10
number_to_guess = random.randint(1, 10)
attempts = 0  # Keep track of the number of attempts

while True:
    try:
        # Get the player's guess
        guess = int(input("Enter your guess (1-10): "))
        attempts += 1

        # Check the guess
        if guess < number_to_guess:
            print("Too low! Try again.")
        elif guess > number_to_guess:
            print("Too high! Try again.")
        else:
            print(f"Congratulations! You guessed the number {number_to_guess} in {attempts} attempts.")
            break
    except ValueError:
        print("Invalid input! Please enter a number between 1 and 10.")
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load a built-in GeoDataFrame of US states
url = "https://raw.githubusercontent.com/PublicaMundi/MappingAPI/master/data/geojson/us-states.json"
us_states = gpd.read_file(url)

# Filter for Missouri
missouri = us_states[us_states['name'] == 'Missouri']

# Plot Missouri
fig, ax = plt.subplots(figsize=(8, 8))
missouri.plot(ax=ax, color='lightblue', edgecolor='black')

# Add a title and labels
ax.set_title("Map of Missouri", fontsize=16, fontweight='bold')
ax.set_xlabel("Longitude", fontsize=12)
ax.set_ylabel("Latitude", fontsize=12)

# Show the plot
plt.show()
```

## Plot the track of Hurricane Katrina

From the cartopy documentation: https://scitools.org.uk/cartopy/docs/latest/gallery/lines_and_polygons/hurricane_katrina.html#sphx-glr-gallery-lines-and-polygons-hurricane-katrina-py

```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import shapely.geometry as sgeom

import cartopy.crs as ccrs
import cartopy.io.shapereader as shpreader


def sample_data():
    """
    Return a list of latitudes and a list of longitudes (lons, lats)
    for Hurricane Katrina (2005).

    The data was originally sourced from the HURDAT2 dataset from AOML/NOAA:
    https://www.aoml.noaa.gov/hrd/hurdat/newhurdat-all.html on 14th Dec 2012.

    """
    lons = [-75.1, -75.7, -76.2, -76.5, -76.9, -77.7, -78.4, -79.0,
            -79.6, -80.1, -80.3, -81.3, -82.0, -82.6, -83.3, -84.0,
            -84.7, -85.3, -85.9, -86.7, -87.7, -88.6, -89.2, -89.6,
            -89.6, -89.6, -89.6, -89.6, -89.1, -88.6, -88.0, -87.0,
            -85.3, -82.9]

    lats = [23.1, 23.4, 23.8, 24.5, 25.4, 26.0, 26.1, 26.2, 26.2, 26.0,
            25.9, 25.4, 25.1, 24.9, 24.6, 24.4, 24.4, 24.5, 24.8, 25.2,
            25.7, 26.3, 27.2, 28.2, 29.3, 29.5, 30.2, 31.1, 32.6, 34.1,
            35.6, 37.0, 38.6, 40.1]

    return lons, lats


def main():
    fig = plt.figure()
    # to get the effect of having just the states without a map "background"
    # turn off the background patch and axes frame
    ax = fig.add_axes([0, 0, 1, 1], projection=ccrs.LambertConformal(),
                      frameon=False)
    ax.patch.set_visible(False)

    ax.set_extent([-125, -66.5, 20, 50], ccrs.Geodetic())

    shapename = 'admin_1_states_provinces_lakes'
    states_shp = shpreader.natural_earth(resolution='110m',
                                         category='cultural', name=shapename)

    lons, lats = sample_data()

    ax.set_title('US States which intersect the track of '
                 'Hurricane Katrina (2005)')

    # turn the lons and lats into a shapely LineString
    track = sgeom.LineString(zip(lons, lats))

    # buffer the linestring by two degrees (note: this is a non-physical
    # distance)
    track_buffer = track.buffer(2)

    def colorize_state(geometry):
        facecolor = (0.9375, 0.9375, 0.859375)
        if geometry.intersects(track):
            facecolor = 'red'
        elif geometry.intersects(track_buffer):
            facecolor = '#FF7E00'
        return {'facecolor': facecolor, 'edgecolor': 'black'}

    ax.add_geometries(
        shpreader.Reader(states_shp).geometries(),
        ccrs.PlateCarree(),
        styler=colorize_state)

    ax.add_geometries([track_buffer], ccrs.PlateCarree(),
                      facecolor='#C8A2C8', alpha=0.5)
    ax.add_geometries([track], ccrs.PlateCarree(),
                      facecolor='none', edgecolor='k')

    # make two proxy artists to add to a legend
    direct_hit = mpatches.Rectangle((0, 0), 1, 1, facecolor="red")
    within_2_deg = mpatches.Rectangle((0, 0), 1, 1, facecolor="#FF7E00")
    labels = ['State directly intersects\nwith track',
              'State is within \n2 degrees of track']
    ax.legend([direct_hit, within_2_deg], labels,
              loc='lower left', bbox_to_anchor=(0.025, -0.1), fancybox=True)

    plt.show()


if __name__ == '__main__':
    main()
```

```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import shapely.geometry as sgeom

import cartopy.crs as ccrs
import cartopy.io.shapereader as shpreader

# From wikipedia, transformed with ChatGPT:
# https://simple.wikipedia.org/wiki/List_of_U.S._states_by_population_density
# Retrieved 1/13/2025
population_density = {
    "District of Columbia": 10588.8,
    "New Jersey": 1210.1,
    "Rhode Island": 1017.1,
    "Massachusetts": 858.0,
    "Connecticut": 742.6,
    "Maryland": 610.8,
    "Delaware": 475.1,
    "New York": 417.0,
    "Florida": 364.6,
    "Pennsylvania": 285.5,
    "Ohio": 283.2,
    "California": 246.1,
    "New England": 233.2,
    "Illinois": 232.0,
    "Hawaii": 218.6,
    "Virginia": 209.2,
    "North Carolina": 202.6,
    "Indiana": 183.4,
    "Michigan": 175.0,
    "Georgia": 173.7,
    "South Carolina": 158.8,
    "Tennessee": 157.5,
    "New Hampshire": 147.8,
    "Kentucky": 111.3,
    "Louisiana": 107.1,
    "Wisconsin": 106.0,
    "Washington": 104.9,
    "Texas": 101.2,
    "Alabama": 95.4,
    "Midwest": 90.0,
    "Missouri": 87.9,
    "West Virginia": 77.1,
    "Minnesota": 68.1,
    "Vermont": 68.0,
    "Mississippi": 63.7,
    "Arizona": 58.3,
    "Arkansas": 56.9,
    "Oklahoma": 56.1,
    "Iowa": 55.3,
    "Colorado": 50.8,
    "Maine": 43.1,
    "West": 42.4,
    "Oregon": 40.9,
    "Kansas": 35.4,
    "Utah": 35.3,
    "Nevada": 25.4,
    "Nebraska": 24.3,
    "Idaho": 19.5,
    "New Mexico": 17.2,
    "South Dakota": 11.1,
    "North Dakota": 10.5,
    "Montana": 7.0,
    "Wyoming": 6.0,
    "Alaska": 1.3
}


def main():
    fig = plt.figure()
    # Set up the map without background and axes frame
    ax = fig.add_axes([0, 0, 1, 1], projection=ccrs.LambertConformal(), frameon=False)
    ax.patch.set_visible(False)
    ax.set_extent([-125, -66.5, 20, 50], ccrs.Geodetic())

    # Load the shapefile
    shapename = 'admin_1_states_provinces_lakes'
    states_shp = shpreader.natural_earth(resolution='110m', category='cultural', name=shapename)

    ax.set_title('US States by Population Density')

    def colorize_state(record):
        statename = record.attributes['name']
        density = population_density[statename]
        if density > 600:
            facecolor = (0, 0, 0)
        else:
            facecolor = (1-density/600, 1-density/600, 1-density/600)
            
        return {"facecolor": facecolor, "edgecolor": "black"}

    # Iterate through the records
    for record in shpreader.Reader(states_shp).records():
        ax.add_geometries(
            [record.geometry],  # Use the geometry from the record
            ccrs.PlateCarree(),
            **colorize_state(record)  # Apply the styling based on the record
        )

    plt.show()


if __name__ == '__main__':
    main()
```
