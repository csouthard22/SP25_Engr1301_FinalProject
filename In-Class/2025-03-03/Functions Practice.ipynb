```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def is_even(n):
    if n % 2 == 0: return True
    else: return False

print(is_even(5))
print(is_even(8))
```

```python
def max_three(a,b,c):
    if a > b:
        if a > c: return a
        else: return c
    elif b > a:
        if b > c: return b
        else: return c
    else: return c

print(max_three(120,25,89))
```

```python
def median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    
    if n % 2 == 0:
        middle1 = sorted_numbers[n // 2 - 1]
        middle2 = sorted_numbers[n // 2]
        return (middle1 + middle2) / 2
    else:
        return sorted_numbers[n // 2]

print(median([4,2,5,8,14,18,35,245,1]))
```

```python
def is_palindrome(text):
    text = text.lower().replace(" ", "")
    return text == text[::-1]

print(is_palindrome("racecar"))
print(is_palindrome("hello"))
print(is_palindrome("taco cat"))
```

```python
fig, ax = plt.subplots()

def draw_circle(x, y, r=1):
    time = [i for i in np.arange(0,2*np.pi+.001,.001)]
    xval = [x + r * np.cos(t) for t in time]
    yval = [y + r * np.sin(t) for t in time]
    plt.plot(xval,yval)
    ax.set_aspect('equal')

draw_circle(3,4)
draw_circle(3,4,2)

plt.show()
```

```python
def circles_intersect(x1,y1,x2,y2,plot=False,r1=1,r2=2):
    if plot == True:
        draw_circle(x1,y1,r1)
        draw_circle(x2,y2,r2)
    
    # Calculate the distance between centers
    distance = np.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    
    # Check if circles intersect
    sum_radii = r1 + r2
    diff_radii = abs(r1 - r2)
    
    # Circles intersect if distance is less than sum of radii and greater than absolute difference of radii
    intersect = diff_radii < distance < sum_radii
    
    # Circles touch if distance equals sum of radii or difference of radii
    touch = np.isclose(distance, sum_radii) or np.isclose(distance, diff_radii)
    
    if plot:
        plt.title(f'Circles {'intersect' if intersect or touch else 'do not intersect'}')
        plt.show()
        
    return intersect or touch

circles_intersect(1,2,5,6,True,4)
```
