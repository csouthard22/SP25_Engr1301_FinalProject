```python
import matplotlib.pyplot as plt
from random import randint 
import numpy
```

# Plotting Probabilities of Dice Combinations

```python
from matplotlib.ticker import PercentFormatter

TRIALS = 10000
def roll():
    return randint(1,6) + randint(1,6)

x = [i for i in range(2,13)]
mathematical = [1/36, 2/36, 3/36, 4/36, 5/36, 6/36, 5/36, 4/36, 3/36, 2/36, 1/36]
empirical = []

for xval in x:
    sum = 0
    for i in range(1,TRIALS):
        if roll() == xval: sum +=1
    empirical.append(sum/TRIALS)

fig, ax = plt.subplots()
ax.stem(x, mathematical, 'r')
ax.bar(x, empirical, color='b')
plt.grid()
plt.title('The Mathematical Probabilities of Rolling Numbers with two Dice')
plt.xlabel('Number Rolled')
plt.ylabel('Mathematical Probability')
ax.set_xticks(x)
ax.yaxis.set_major_formatter(PercentFormatter(1))
plt.show()

# stem represents mathematically certain values, bars represent empirical data as found by calculation
# as TRIALS increases, the bars are closer to the true values
```

# Throwing Ball Animation

```python
from IPython.display import clear_output

def height(t,vo,a):
    return (vo*t) - 0.5*(a*t**2)

V_INITIAL = 43.81
G_ACCEL = 9.8
ENDTIME = 2*V_INITIAL/G_ACCEL

times = [x*ENDTIME/100 for x in range(0,101)]
heights = [height(t,V_INITIAL,G_ACCEL) for t in times]

fig, ax = plt.subplots()
for h in heights:
    plt.clf() # Clear plot
    clear_output(wait=True)

    ax = plt.gca()
    ax.scatter(0,h)
    ax.axis([-1,1,0,100])

    plt.show()
```
