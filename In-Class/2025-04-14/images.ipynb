# Image Data in Python

Most of the plotting we've done relies on specifying objects and ignoring how the underlying images are created (rendered) onto the screen. For example, a line plot specifies lines, a bar plot specifies bars, a scatter plot specifies points, and so on.

We haven't needed to get access to the actual image data itself, but sometimes that can be necessary. For example:

* Our data is a set of pictures of cancer cells on microscope slides
* We want to generate an image that is not a plot

As data, images are stored as matrices. You can think of these as two dimensional matrices that store a color at every row-column location, or as three dimensional matrices that have a row, column, and then a third dimension with three spots for red, green, and blue color values.

The code block below creates a black image with three colored pixels. 


```python
import matplotlib.pyplot as plt
import numpy as np

# Create an empty image matrix 
width = 50
height = 50
# Creates a 50x50x3 array, so 50x50 pixels with three color channels
# for Red, Green, and Blue. All of these start as zero, so the image
# starts out completely black.
canvas = np.zeros( (height, width, 3) )

# We can add colors to pixels one at a time
# Note the image coordinate system might not be what you expect. Matrices
# are accessed in [ ROW, COLUMN ] order. The first coordinate controls
# the pixel row and starts at zero at the top. The the second controls 
# pixel column and starts at zero on the left.
canvas[25, 25] = [1, 0, 0]  # Red pixel at center
canvas[5, 5] = [0, 1, 0]  # Green pixel at top left
canvas[25, 45] = [0, 0, 1]  # Blue pixel at right edge

# Show the image
fig, ax = plt.subplots()
ax.imshow(canvas)

plt.show()
```

# Image Coordinates

Images matrices are accessed in as matrices, not as plots. In an x-y plot, we use (x,y) coordinates, and the origin is usually the center of the plot or the bottom left corner. 

With an image matrix, we access them by (row, column) coordinates, where numbering starts at the top left. The top row is row zero, and the leftmost column in column zero. It can sometimes be hard to wrap your mind around different coordinate systems on the fly. The next image demonstrates this by drawing a little house.

```python
# Create an empty image matrix 
width = 7
height = 7
# Note we use np.ones() this time, so our image starts out white
canvas = np.ones( (height, width, 3) )

# Draw the ground
canvas[6, 0] = [0, 0, 0]
canvas[6, 1] = [0, 0, 0]
canvas[6, 2] = [0, 0, 0]
canvas[6, 3] = [0, 0, 0]
canvas[6, 4] = [0, 0, 0]
canvas[6, 5] = [0, 0, 0]
canvas[6, 6] = [0, 0, 0]

# Draw left wall
canvas[5, 2] = [0, 0, 0]
canvas[4, 2] = [0, 0, 0]

# Draw right wall
canvas[5, 4] = [0, 0, 0]
canvas[4, 4] = [0, 0, 0]

# Draw roof horizontal
canvas[3, 1] = [0, 0, 0]
canvas[3, 2] = [0, 0, 0]
canvas[3, 3] = [0, 0, 0]
canvas[3, 4] = [0, 0, 0]
canvas[3, 5] = [0, 0, 0]

# Draw roof middle
canvas[2, 2] = [0, 0, 0]
canvas[2, 3] = [0, 0, 0]
canvas[2, 4] = [0, 0, 0]

# Draw roof peak
canvas[1, 3] = [0, 0, 0]

# Show the image
fig, ax = plt.subplots()
ax.imshow(canvas)

plt.show()
```

# Manipulating Image Data

Images are loaded as Numpy matrices just like before, but usually with hundreds or thousands of rows and columns. We can't modify that data on a pixel-by-pixel basis, but we can write some code.

For example, the following code replaces a big chunk of the photo with a red block.

```python
# Load the image
img = mpimg.imread('brookings.jpg') 

# Make a writable copy
copy = np.array(img, copy=True)

start_row = 400
start_col = 400
width = 300
height = 200

copy[400:700, 400:600] = [255, 0, 0]  # Change a block of pixels to red

# Display the image
plt.imshow(copy)
plt.axis('off')  # Optional: turn off axes
plt.show()
```
