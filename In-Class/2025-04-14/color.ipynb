# Creating Custom Colors

So far, most of the colors we've used in our code in class have been specified with a color name, like `red`, `green`, or `blue`. 

You can find a fuller list of colors here:

https://matplotlib.org/stable/gallery/color/named_colors.html

However, sometimes we don't want a named color, we want to compute a color. For example, we might want to:

* Color states according to their population density
* Color soccer players according to their height
* Color stock investment returns based on how positive or negative they are

The first cell below plots three shapes with named colors from the link above. 

The second cell plots three shapes based on custom colors? Can you spot the difference?

```python
# Plot three figures based on named color
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a new figure and axes
fig, ax = plt.subplots()

# Add a square
square = patches.Rectangle((0.3, 0.1), 0.3, 0.5, angle=20, facecolor='darkorchid')
ax.add_patch(square)

# Add a circle
circle = patches.Circle((0.75, 0.50), 0.15, facecolor='goldenrod')
ax.add_patch(circle)

# Add a triangle (as a polygon)
triangle = patches.Polygon([[0.5, 0.9], [0.3, 0.7], [0.7, 0.7]], facecolor='steelblue')
ax.add_patch(triangle)

# Display the plot
plt.show()
```

```python
# Plot three figures based on CUSTOM colors
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a new figure and axes
fig, ax = plt.subplots()

# Add a square
square = patches.Rectangle((0.3, 0.1), 0.3, 0.5, angle=20, facecolor=(1.0,0.0,0.0))
ax.add_patch(square)

# Add a circle
circle = patches.Circle((0.75, 0.50), 0.15, facecolor=(0.0,1.0,0.0))
ax.add_patch(circle)

# Add a triangle (as a polygon)
triangle = patches.Polygon([[0.5, 0.9], [0.3, 0.7], [0.7, 0.7]], facecolor=(0.0,0.0,1.0))
ax.add_patch(triangle)

# Display the plot
plt.show()
```

# RGB Color Space

The Red Green Blue color space specifies colors as a combination of those three primary colors. 

Wait a second, didn't my art teacher tell me that the primary colors are Red, Yellow, and Blue? Well, it depends. Computer displays *emit light*, while paints, inks, and dyes all *reflect light* from the environment. A computer display is made up of many tiny light emitters, so it creates its' own colors, while a painting reflects the light that shines on it.

# Additive and Subtractive Color

In practice, this gives us two different primary color systems. The RGB system is an *additive* or *emissive* color system, while paints, inks, dyes, etc. are captured by a *subtractive* or *reflective* color system. 

Let's describe the physics of what is going on. A computer monitor has many tiny red, green, and blue light emitters packed together. When you look at the red rectangle above, the monitor turns on the red light emitters and turns off the green and blue emitters. The rectangle is red because only red light is generated at that spot.

However, if we painted a blue triangle and hung it on the wall, we have to shine a light on it to see it. If we shine a white light on the painting, the triangle looks blue because it absorbs every color except blue, and reflects just the blue light to our eyes. 

In an emissive color system, the color red is just the color red being emitted. In a reflective color system, everything that is not red is being absorbed. 

# Color Combinations

The primary emissive color system is the RGB system. This is used any time light emitting elements are used, such as in computing, but also in contexts like theatrical lighting.



Red and green combine to make yellow. Red and blue combine to make magenta. Blue and green combine to make cyan.

The colors in the second plot above are specified as a set of three numbers in the RGB color space. The first number is the Red component, 
the second number is the Green component, and the third is Blue. 

These colors mix in the same way as the image above. See the code below.

```python
# Plot the eight standard colors - red, green, blue,
# cyan, magenta, yellow, and white.
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a new figure and axes
fig, ax = plt.subplots()
ax.set_facecolor('black')

# Add red circle
circle = patches.Circle((0.5, 1.0), 0.15, facecolor=(1.0,0.0,0.0))
ax.add_patch(circle)

# Add green circle
circle = patches.Circle((0.0, 0.2), 0.15, facecolor=(0.0,1.0,0.0))
ax.add_patch(circle)

# Add blue circle
circle = patches.Circle((1.0, 0.2), 0.15, facecolor=(0.0,0.0,1.0))
ax.add_patch(circle)

# Add yellow circle
circle = patches.Circle((0.15, 0.6), 0.15, facecolor=(1.0,1.0,0.0))
ax.add_patch(circle)

# Add magenta circle
circle = patches.Circle((0.85, 0.6), 0.15, facecolor=(1.0,0.0,1.0))
ax.add_patch(circle)

# Add cyan circle
circle = patches.Circle((0.5, 0.1), 0.15, facecolor=(0.0,1.0,1.0))
ax.add_patch(circle)

# Add white circle
circle = patches.Circle((0.5, 0.5), 0.15, facecolor=(1.0,1.0,1.0))
ax.add_patch(circle)

# Display the plot
plt.axis('equal')
plt.show()
```

# Zero, One, 255?

The two dominant ways to represent RGB colors are either on a decimal scale from zero to one, or on an integer scale zero to 255. Why 255? Remember that a single byte of data, a single 8-bit integer, can only store values between 0 and 255. If we use one byte for each of Red, Green, and Blue, then we get 256 possible color levels from zero to 255. 

For example, here's a color table and color picker that uses 0 to 255.

https://www.rapidtables.com/web/color/RGB_Color.html

Likewise, here's a color picker that uses 0.0 to 1.0.

https://rgbcolorpicker.com/0-1

If you ever want to convert between the 0-255 scale and the 0.0-1.0 scale, just divide the number by 255!

# White, Black, and Grey

How do we make these special colors? Remember that white light is the presence of color components, so white is `(1.0, 1.0, 1.0)`. Likewise, black is the absence of all color components, so it is `(0.0, 0.0, 0.0)`. 

Grey is what you get when the three color components have equal levels between those two extremes. See the code below for an example.

```python
# Plot three shades of grey
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a new figure and axes
fig, ax = plt.subplots()
ax.set_facecolor('black')

# Add dark grey
circle = patches.Circle((0.2, 0.5), 0.15, facecolor=(0.25,0.25,0.25))
ax.add_patch(circle)

# Add middle grey
circle = patches.Circle((0.5, 0.5), 0.15, facecolor=(0.5,0.5,0.5))
ax.add_patch(circle)

# Add light grey
circle = patches.Circle((0.8, 0.5), 0.15, facecolor=(0.75,0.75,0.75))
ax.add_patch(circle)

# Display the plot
plt.axis('equal')
plt.show()
```

# Color Ranges

We can visualize the gradient, the set of all colors, by varying these continuously. You might have seen gradients like this before! 

A flat image is only two dimensional, so we can only do gradients like this with two colors at a time. In the first image we have red and green, where the top row is zero red and the bottom row is 100% red, and the left column is zero green and the right column is 100% green. Thus, we get black in the top left, pure red in the bottom left, pure green in the top right, and pure yellow in the bottom right.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create an empty image matrix 
width = 256
height = 256
rg_canvas = np.zeros( (height, width, 3) )
rb_canvas = np.zeros( (height, width, 3) )
gb_canvas = np.zeros( (height, width, 3) )

for row in range(height):
    for col in range(width):
        rg_canvas[row, col] = [ row/255.0, col/255.0, 0 ]
        
for row in range(height):
    for col in range(width):
        rb_canvas[row, col] = [ row/255.0, 0, col/255.0 ]
        
for row in range(height):
    for col in range(width):
        gb_canvas[row, col] = [ 0, row/255.0, col/255.0 ]

# Show the first image
fig, ax = plt.subplots()
ax.imshow(rg_canvas)
plt.title("Red (row) / Green (column) Gradient")
plt.show()

# Show the second image
fig, ax = plt.subplots()
ax.imshow(rb_canvas)
plt.title("Red (row) / Blue (column) Gradient")
plt.show()

# Show the third image
fig, ax = plt.subplots()
ax.imshow(gb_canvas)
plt.title("Green (row) / Blue (column) Gradient")
plt.show()
```
