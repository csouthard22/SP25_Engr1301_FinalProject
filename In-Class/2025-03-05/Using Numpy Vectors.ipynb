```python
import numpy as np
import matplotlib.pyplot as plt
```

# Building Vectors

```python
v = np.arange(100,1600,100)
print(v)
```

```python
v1 = np.arange(1,20)
v2 = np.arange(20, 0, -1)
v = np.append(v1, v2)
print(v)
```

```python
v = np.arange(1,-2,-2)
v1 = v.copy()
for i in np.arange(4):
    v = np.append(v,v1)
print(v)
```

```python
v = np.tile([1, -1], 5)
print(v)
```

```python
v = np.array([2**i for i in np.arange(10)])
print(v)
```

```python
v = np.array([1/(2**i) for i in np.arange(10)])
print(v)
```

```python
v = np.array([1/(2**i) for i in np.arange(20)])

print(np.sum(v))
print(np.cumsum(v))
```

# Plotting Data

```python
t = np.arange(0, np.pi*2, 0.01)

fig, ax = plt.subplots()
ax.plot(np.cos(t), np.sin(t))
plt.show()
```

```python
t = np.arange(0, np.pi*4, 0.01)

fig, ax = plt.subplots()
ax.plot(t*np.cos(t), t*np.sin(t))
plt.show()
```

```python
K = .4
L = 3

t = np.arange(0, 100, 0.01)
x = (1-K)*np.cos(t)+L*K*np.cos(((1-K)/K)*t)
y = (1-K)*np.sin(t)+L*K*np.sin(((1-K)/K)*t)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_aspect('equal')
ax.grid(True)
plt.show()
```
